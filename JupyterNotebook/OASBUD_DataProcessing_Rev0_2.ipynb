```python
from scipy.fftpack import fft
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
from tqdm import tnrange, tqdm_notebook
from scipy.io import loadmat
from scipy import signal,ndimage
import matplotlib.pyplot as plt
import scipy.io as sio
import numpy as np
import os
import random
import time
import timeit
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
# load the OASBUD raw data from google drive
# The data is also available on website: https://zenodo.org/record/545928#.XY0eSkZKiic
origindata = loadmat("C:/Users/wangb/Documents/MATLAB/ultrasonic_image_nn/OASBUD_for_Python.mat")
subject = ["id", "rf1", "rf2", "roi1", "roi2", "birads", "class"]

tmpnames = origindata["id"]
tmprf1 = origindata["rf1"]
tmprf2 = origindata["rf2"]
tmproi1 = origindata["roi1"]
tmproi2 = origindata["roi2"]
tmptypes = origindata["class"]
tmpbirads = origindata["birads"]

names = []
rf = []
rfenvelope = []
roi = []
types = []
birads = []
timeaxis = []
roiedge = []
rfrfft = []
freqaxis = []

for index in range(0,100):
    names.append(tmpnames[0,index][0]+'_0')
    names.append(tmpnames[0,index][0]+'_1')
    
    rf.append(tmprf1[0,index][20:,:])
    rf.append(tmprf2[0,index][20:,:])
    
    roi.append(tmproi1[0,index][20:,:])
    roi.append(tmproi2[0,index][20:,:])
    
    types.append(tmptypes[0,index][0][0])
    types.append(tmptypes[0,index][0][0])
    
    birads.append(tmpbirads[0,index][0])
    birads.append(tmpbirads[0,index][0])
    
del tmprf1,tmprf2,tmproi1,tmproi2,tmptypes,tmpbirads

#Generating Time axis for the data
Fs = 40e6
T = 1/Fs

for index in range(0,len(rf)):
    N = rf[index].shape[0]
    t = T * N
    timeaxisbuf = np.linspace(0,t,N)*1e6 #time axis in us
    timeaxis.append(timeaxisbuf)

# Find roi edge of the region of interest
for items in roi:
    edgebuf = ndimage.sobel(items,output="bool",cval=100)
    roiedge.append(edgebuf)

# Find rf envelope
for items in rf:
    envelope = 20 * np.log10(abs(signal.hilbert(items,axis=0)))
    rfenvelope.append(envelope)

# Compute rfft of the signal
for items in rf:
    rfftresult = np.fft.rfft(items,axis=0)
    rfrfft.append(rfftresult)

# Compute frequency axis
for items in rfrfft:
    axisbuffer = np.linspace(0,Fs/2,items.shape[0])/1e6
    freqaxis.append(axisbuffer)
    
```

```python
randindex = random.randint(0,len(rf))

testtime = timeaxis[randindex]
testrf = rf[randindex]
testroi = roi[randindex]
testroiedge = roiedge[randindex]
testbirads = birads[randindex]
testclass = types[randindex]
testname = names[randindex]
testrfenvelope = rfenvelope[randindex]
testfreq = freqaxis[randindex]
testrfrfft = rfrfft[randindex]

#plot the contour 
fig = plt.figure(figsize = (18,10))
ax = fig.subplots()
testplot = testrfenvelope * np.invert(testroiedge)
xaxis = testtime
yaxis = np.linspace(0,testplot.shape[1]-1,num=testplot.shape[1])
_ = ax.contourf(yaxis,xaxis,testplot,max=np.amax(testrfenvelope),min=np.amin(testrfenvelope),origin="lower")
ax.invert_yaxis()
plt.show()
```

```python
#find the xaxis range of the region of interest
indexmin = testroi.shape[1]
indexmax = -1
for i in range(0,testroi.shape[1]-1):
    if np.any(testroi[:,i]) and i < indexmin:
        indexmin = i
    if np.any(testroi[:,i]) and i > indexmax:
        indexmax = i

# get a random index with target echo in it.
index2 = random.randint(indexmin,indexmax)

_ = plt.figure(figsize = (15,4))
_ = plt.subplot(211)
_ = plt.plot(testtime,testrf[:,index2])
_ = plt.xlabel('Time [us]', fontsize=16)
_ = plt.subplot(212)
_ = plt.plot(testtime,testroi[:,index2])
_ = plt.xlabel('Time [us]', fontsize=16)
```

```python
#SSP Processing
freq_axis = np.linspace(0,20e6,testrfrfft.shape[0])

result = gaussian(freq_axis,10e6,10e5)
maxval = float(np.linalg.norm(testrfrfft.real))

#Generate Gaussian kernel for filter
gaussianstep = 5
listofmu = []
startfreq = 1.5e6
sigma = 3e5
endfreq = 12e6

for items in freq_axis:
    if items>=startfreq and items<endfreq:
        listofmu.append(items)


listofmu = listofmu[0::gaussianstep]
listlen = len(listofmu)

kernel_matrix = np.zeros((testrfrfft.shape[0],listlen))

for i in range(listlen):
    kernel_matrix[:,i] = gaussian(freq_axis,listofmu[i],sigma)

plt.figure(figsize=(15,3))
_ = plt.plot(freq_axis[:]/1e6,testrfrfft.imag[:]/maxval*60,color = "y")
_ = plt.plot(freq_axis[:]/1e6,testrfrfft.real[:]/maxval*60,color = "b")
_ = plt.plot(freq_axis[:]/1e6,kernel_matrix[:,:])
_ = plt.xlabel('Frequency [MHz]', fontsize=16)

# Compute SSP result for testsignal
maskedresult = kernel_matrix.transpose()*testrfrfft[:,index2]
maskedresult = maskedresult.transpose()
testssp = np.fft.irfft(maskedresult,axis=0)
```

```python
testsspfreq = np.asarray(listofmu)

plt.figure(figsize=(15,10))
plt.subplot(3,1,1)
#print(frequencylist.shape,time_axis.shape,sspresult.shape)
_ = plt.plot(testtime,testrf[:,index2])
plt.xlim(0,testtime[-1])

plt.subplot(3,1,2)
testsspenvelope = 20 * np.log10(signal.hilbert(testssp,axis=0))
_ = plt.contourf(testtime,testsspfreq/1e6,testsspenvelope.transpose())

plt.subplot(3,1,3)
_ = plt.plot(testtime,testroi[:,index2])
_ = plt.xlabel('Time [us]', fontsize=16)
plt.xlim(0,testtime[-1])

plt.show()
```
